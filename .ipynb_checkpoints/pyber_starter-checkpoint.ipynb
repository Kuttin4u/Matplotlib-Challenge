```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_data = pd.read_csv(city_data_to_load)
ride_data = pd.read_csv(ride_data_to_load)

# Combine the data into a single dataset
city_ride_df = pd.merge(city_data, ride_data, how='right')


# Display the data table for preview
city_ride_df.head()
```

```python
# Obtain the x and y coordinates for each of the three city types
city_type = city_ride_df.groupby('type')
count_types =city_type['type'].count()
#city_type.head()
count_types.head()
```

## Bubble Plot of Ride Sharing Data

```python


avg_fare =city_ride_df['fare'].mean()
rides_per = city_ride_df['ride_id'].count()
drivers = city_ride_df['driver_count'].count()


# Build the scatter plots for each city types
#plt.scatter(city_type,count_types,  marker="o", facecolors="Gold", edgecolors="black")

plt.scatter(rides_per, avg_fare, marker="o", label='Rural',s=10,facecolors="SkyBlue", edgecolors="black")
plt.scatter(rides_per, drivers, marker="o", label='Suburban',s=20,facecolors="Coral", edgecolors="black")
#plt.scatter(rides_per, city, marker="o", label='Urban',s=30,facecolors="Gold", edgecolors="black")

#city_ride_df.plot(kind='scatter', x=[count_types], y=[avg_fare])
    
# Incorporate the other graph properties
plt.title("Ride Sharing Data")
plt.xlabel("Number of Rides per City")
plt.ylabel("Average Fare")
plt.show()

# Create a legend
#legend = plt.legend(loc="lower right", scatterpoints=1, fontsize=8)
#legend.legendHandles[0]._sizes = [10]
#legend.legendHandles[1]._sizes = [10]

# Incorporate a text label regarding circle size

# Save Figure
plt.savefig("../RidesbyCityType.png")
```

```python
# Show plot
plt.show()
```

## Total Fares by City Type

```python
type_percent = city_ride_df.groupby('type')
fares_by_type =type_percent['fare'].sum()
fares_by_type
```

```python
# Calculate Type Percents
labels = ['Rural', 'Suburban', 'Urban']
fares = [4327.93, 19356.33, 39854.38]
colors = ['SkyBlue', 'Coral', "Gold"]
explode = (0,0,0.1)

# Build Pie Chart
plt.pie(fares, explode=explode, labels = labels, colors=colors, autopct="%1.1f%%", shadow=True,startangle=140)

# Save Figure
plt.savefig("../RidesbyTypePercent.png")
```

```python
# Show Figure
plt.show()
```
